## Step 6: Configure your Git identity

Before Git can create commits, it must know who is making them.

Run these commands in the terminal and replace the sample values with your own information:

```bash
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```

### Optional commands from the chapter
These are useful and match the chapter discussion of editor and default branch setup:

```bash
git config --global core.editor "code --wait"
git config --global init.defaultBranch main
```

### Why this matters
Every commit stores author information.  
This is important for attribution, collaboration, and project history.


## Step 7: Verify your Git configuration

Run:

```bash
git config --global --list
```

### Check that you can see
- `user.name`
- `user.email`

If you also ran the optional setup, you may also see:
- `core.editor`
- `init.defaultbranch`

### Important note
If your name or email is incorrect, run the `git config --global` command again with the correct value.


## Step 8: Initialize the repository

Now turn the current folder into a Git repository.

In the terminal, run:

```bash
git init
```

### What should happen
Git creates a hidden folder named `.git`.

This hidden folder stores:
- repository history,
- configuration,
- references to commits and branches.

### Why this matters
This is the step that changes an ordinary folder into a **Git repository**.


## Step 9: Check repository status

Immediately after `git init`, run:

```bash
git status
```

### Read the output carefully
You should see information about:
- the current branch,
- files that are **untracked**,
- and what Git expects you to do next.

### Important idea from the chapter
`git status` acts like your **compass** or **radar**.  
Use it before and after important Git commands so you always know what changed.


## Step 10: Stage your first file

Add `README.md` to the staging area:

```bash
git add README.md
git status
```

### What should change
Before `git add`, the file is **untracked**.  
After `git add`, the file should appear under **Changes to be committed**.

### What the staging area means
The staging area is where you choose exactly what will go into the next snapshot.


## Step 11: Create your first commit

Now save the staged file permanently in repository history:

```bash
git commit -m "Add initial README for Chapter 9 lab"
```

Then check the repository again:

```bash
git status
```

### What you should see
Git should report that there is **nothing to commit** if no other changes exist.

### What this teaches
A **commit** is a named snapshot of your project at a specific moment.


## Step 12: Create a second file for comparison

Now create another file named `notes.txt`.

Create `notes.txt` in VS Code and type the following content:


````python

This file is used to compare staged and unstaged changes.
````

### What to do now
1. Save `notes.txt`.  
2. Run:

```bash
git status
```

### What to observe
At this point, `notes.txt` should appear as **untracked** because it has not been staged yet.


## Step 13: Compare tracked, staged, and untracked work

This step helps make the chapter ideas clearer.

Run the following commands:

```bash
git add notes.txt
git status
```

### Explanation
- a file is **untracked** when Git has never saved it before,
- a file is **staged** when it is selected for the next commit,
- a file is **tracked** after it has been committed at least once.

This is one of the most important concepts in the chapter.


## Step 14: Commit the second file

Create a second commit:

```bash
git commit -m "Add notes file for staging practice"
```

Then run:

```bash
git status
```

### What this teaches
Git lets you create separate snapshots for separate changes.  
That is much better than saving many files with names like `final_v2_REAL_FINAL`.


## Step 15: Modify a tracked file

Open `README.md` again and add this new section at the bottom:


````python

## Reflection
Git helps track changes without creating many duplicate filenames.
````

### What to do now
1. Save `README.md`.  
2. Run:

```bash
git status
```

### Important note
Now `README.md` should appear as **modified**, not untracked.

This difference matters:
- **untracked** = Git has never committed this file before
- **modified** = Git already knows the file, but you changed it since the last commit


## Step 16: Review the exact change with `git diff`

Run:

```bash
git diff
```

### What this shows
`git diff` displays the line-by-line difference between:
- the last committed version,
- and the current working version.

### What to look for
You should see the new `Reflection` section added to `README.md`.


## Step 17: Stage and commit the updated file

Now save the new version in Git history:

```bash
git add README.md
git commit -m "Update README with reflection section"
```

Then verify the result:

```bash
git status
```

### What this teaches
The normal local Git workflow is:

```text
edit -> status -> add -> commit -> status
```


## Step 18: View commit history

Run:

```bash
git log --oneline
```

### What you should see
You should now see a short history with your commit messages, for example:
- Add initial README for Chapter 9 lab
- Add notes file for staging practice
- Update README with reflection section

Each line represents a snapshot in your repository history.

In the VS Code terminal, run:

